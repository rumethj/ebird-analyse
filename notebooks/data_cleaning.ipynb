Load the data

```python

checklist_data_path = "../data/checklists"
checklist_files_list = os.listdir(checklist_data_path)
checklist_df = pd.DataFrame()
for file in checklist_files_list:
    if file.endswith('.tsv'):
        file_path = f"{checklist_data_path}/{file}"
        checklist_df = pd.concat([checklist_df, pd.read_csv(file_path, sep="\t")], ignore_index=True)

print("\n" + "="*50)
print("DataFrame Head:")
print("="*50)
print(checklist_df.head())
print(f"\nShape: {checklist_df.shape}")
print(f"Columns: {list(checklist_df.columns)}")
```
